```python
import pandas as pd

dataset = pd.read_csv("google-play-store.csv")
dataset = dataset.dropna(subset=['genre', 'min_installs'])
```

```python
# Task 1 قسمت التطبيقات حسب الفئه وحسبت المتوسط ورتبت تنازلي
dataset.groupby('genre')['min_installs'].median().sort_values(ascending=False).head(10)
```

```python
# Task 2 رتبت تصاعدي
dataset.groupby('genre')['min_installs'].median().sort_values(ascending=True).head(10)
```

```python
# Task 3 بحسب اكبر تطبيق من حيث عدد التنزيلات عن طريق اكبر رقم والمتوسط وقسمه اكبر رقم علي المتوسط  وطباعه 5 اعداد
max_installs = dataset.groupby('genre')['min_installs'].max()
median_installs = dataset.groupby('genre')['min_installs'].median()

leader_ratio = (max_installs / median_installs).sort_values(ascending=False)

leader_ratio.head(5)
```

```python
# Task 4 حساب المتوسط
dataset['min_installs'].median()
```

```python
# Task 5 اعلي تنزيل من التطبيقات من حيث التطبيقات المدفوعه 
dataset.groupby(dataset['price'] > 0)['min_installs'].median()
```

```python
# Task 6 اي التطبيقات اعلي تقييم ترتيب تنازلي واظهار اعلي 5 تطبيقات 
dataset.groupby('genre')['score'].mean().sort_values(ascending=False).head(5)
```

```python
# Task 7 اي تطبيقات يوجد بها اعلي تفاعل وهعرض 5 تطبيقات منهم ترتيب تنازلي
dataset.sort_values(by='reviews', ascending=False)[['title', 'reviews']].head(6)
```

```python
# Task 8 كم تطبيق يعتمد علي علي الoffers_iap
dataset['offers_iap'].sum()
```

```python
# Task 9 بحسب اجمالي عدد التحميلات وبرتب ترتيب تنازلي وبعرض اول 7 تطبيقات
dataset.groupby('genre')['min_installs'].sum().sort_values(ascending=False).head(7)
```

```python
# Task 10 بجيب تاريخ اقدم واحدث تطبيق من حيث الرفع
dataset['released'] = pd.to_datetime(dataset['released'])
dataset['released'].min(), dataset['released'].max()
```

```python
# Task 11 بجيب متوسط التقييمات
dataset[dataset['reviews'] > 10000]['ratings_per_day'].mean()
```

```python
# Task 12  اعلي 4 تطبيقات من حيث الاسكور
dataset.groupby('genre')['score'].mean().sort_values(ascending=False).head(4)
```

```python
# Task 13 تم الاستعانه بشات وفهمت قد اي التطبيقات بتستخدم طريق ربح مزدوج مثل الاعلانات و الشراء الاوفلاين
both = dataset[(dataset['ad_supported'] == True) &
(dataset['offers_iap'] == True)]
percentage = (len(both) / len(dataset)) * 100
percentage
```

```python
# Task 14 السنه اللي فيها اعلي نسبه نشر تطبيقات 
dataset['release_year'] = dataset['released'].dt.year

dataset['release_year'].value_counts().head(1)
```

```python
# Task 15 بجيب اعلي التطبيقات عليها ريفيو وبشوف الاعلي سكور فيها وبطبعله 5 منهم
def genre_corr(group):
    return group['reviews'].corr(group['score'])

dataset.groupby('genre')[['reviews','score']].apply(genre_corr).head(5)
```

```python
# Task 16 تم الاستعانه ب شات وحاولت افهمه لكن للاسف ماباليد حيله
weighted_rating = dataset.groupby('genre') \
    .apply(lambda x: (x['score'] * x['reviews']).sum() / x['reviews'].sum())
 
weighted_rating.sort_values(ascending=False).head(10)
```

```python
# Task 17 بجيب اعلي نسبه تقييم بالنسبه لاعلي نسبه تفاعل وبعرض منها 5 تطبيقات
dataset['rating_review_ratio'] = dataset['ratings'] / dataset['reviews']

dataset.loc[dataset.groupby('genre')['rating_review_ratio'].idxmax(),['genre', 'title', 'rating_review_ratio']].head(5)
```
